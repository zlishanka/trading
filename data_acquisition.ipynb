```python
import yfinance as yf

# Define the ticker symbol
ticker_symbol = "TSLA"

# Create a ticker object
ticker = yf.Ticker(ticker_symbol)

# Download historical data
#historical_data = ticker.history(start="1990-01-01", end="2022-12-31", interval='1d')
#historical_data = ticker.history(interval="90m", start="2023-06-01", end="2023-07-01")
#ticker.history(interval="1d", start="1990-01-01", end="2022-12-31")
historical_data = ticker.history(period="1200d", interval='1d')
```

```python
len(historical_data)
```

available timeframes: 1m, 2m, 5m, 15m, 30m, 60m, 90m, 1h, 1d, 5d, 1wk, 1mo, 3mo >>>
1m data is only retrievable for the last 7 days, and anything intraday (interval <1d) only for the last 60 days.
